```python
import ctypes as ct
ir_dll_path = 'libirimager.dll'
LibIRImagerDLL = ctypes.CDLL(ir_dll_path)       # Load DLL
xml_path = ctypes.c_char_p(b'./21114001.xml')   # Config path
format_file_path = ctypes.c_char_p(b'./')       # Location of formats.def
log_file_path = ctypes.c_char_p(b'./testlog')   # Where to put log + prefix
camera_serial = ctypes.c_ulong()                # Storage for camera SN
```
